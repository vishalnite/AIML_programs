## P2 - KMeans Clustering
This Python code uses the scikit-learn library to demonstrate k-means clustering and Gaussian Mixture Model (GMM) clustering on the Iris dataset. The Iris dataset is loaded, and a DataFrame is created for both features (sepal and petal dimensions) and target labels. The code then applies k-means clustering with three clusters and plots the original class labels based on petal features alongside the k-means clustering results. Subsequently, a Gaussian Mixture Model with three components is fitted to the data, and the resulting clusters are plotted. The visual comparison of the original classes, k-means clusters, and GMM clusters provides insights into the performance of these clustering techniques. K-means is a centroid-based clustering algorithm that assigns data points to the nearest cluster center, while GMM is a probabilistic model that represents the data as a mixture of Gaussian distributions, allowing for more flexible cluster shapes and capturing the underlying data distribution.

```python
from sklearn import datasets 
from sklearn.cluster import KMeans
import sklearn.metrics as metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# import some data to play with 
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
X.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width'] 
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
# Build the K Means Model
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
plt.figure(figsize=(14,7))
colormap = np.array(['red', 'lime', 'black'])
```

```python
# Plot the Original Classifications using Petal features
plt.subplot(1, 3, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40) 
plt.title('Real')
plt.xlabel('Petal Length') 
```

```python
# Plot the Models Classifications
plt.subplot(1, 3, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('K-Means Clustering')
plt.xlabel('Petal Length') 
plt.ylabel('Petal Width')
# plt.ylabel('Petal Width')
```

```python
# General EM for GMM
# from sklearn import preprocessing 
from sklearn.mixture import GaussianMixture 

gmm= GaussianMixture(n_components=3, random_state=0).fit(X)
z=gmm.predict(X)
```

```python
plt.subplot(1, 3, 3)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[z])
plt.title('GMM Clustering')
plt.xlabel('Petal Length') 
plt.ylabel('Petal Width')
```
